**1. Id**
- Explanation: A unique identifier for each row in the dataset. This is generally just a technical reference and not used for classification.
- Usefulness: Not needed for classification.

**2. OrgId**
- Explanation: Represents the organization where the incident or alert occurred. Different organizations might have different risk profiles.
- Usefulness: Could be useful when analyzing incidents across multiple organizations, but might not be critical for all models.

**3. IncidentId**
- Explanation: A unique identifier for each incident, grouping together all the alerts related to a single security incident.
- Usefulness: Not needed for classification itself but useful for grouping data.

**4. AlertId**
- Explanation: A unique identifier for each alert. Each incident might trigger multiple alerts.
- Usefulness: Primarily for tracking rather than classification.

**5. Timestamp**
- Explanation: The exact date and time when the alert was generated.
- Usefulness: Useful for understanding time-based patterns in incidents, such as spikes during certain hours.

**6. DetectorId**
- Explanation: Identifies the detection system that generated the alert. Different detectors might be better at identifying certain types of threats.
- Usefulness: Useful for analyzing if certain detectors are more likely to generate false positives or if they are better at detecting specific threats.

**7. AlertTitle**
- Explanation:  A numeric code that likely corresponds to a specific type of alert.
- Usefulness: Could be important for understanding the nature of the alert.

**8. Category**
- Explanation: The type of threat or behavior detected, such as "LateralMovement" or "CommandAndControl."
- Usefulness: Highly useful for classification as it directly describes the type of threat, which is crucial for determining the severity or nature of the incident.

**9. MitreTechniques**
- Explanation: Lists MITRE ATT&CK techniques used by the attackers, which are standardized ways of describing what the attackers did.
- Usefulness: Very useful for classification, especially when predicting the method or severity of the attack.

**10. IncidentGrade**
- Explanation: The severity or confidence level assigned to the incident (e.g., "BenignPositive," "FalsePositive").
- Usefulness: Important for classification, especially when distinguishing between benign and malicious incidents.

**11. ActionGrouped**
- Explanation: A broad category of the action taken in response to the alert.
- Usefulness: Useful for understanding response patterns, which could be relevant in certain classification models.

**12. ActionGranular**
- Explanation: A more detailed description of the action taken in response to the alert.
- Usefulness: Similar to ActionGrouped but more detailed, helping to understand what specific steps were taken.

**13. EntityType**
- Explanation: Describes what kind of entity was involved in the incident (e.g., user, device, file).
- Usefulness: Useful for classification when determining the type of attack target.

**14. EvidenceRole**
- Explanation: Indicates the role of the evidence in the investigation (e.g., Related, Impacted).
- Usefulness: May help in determining the reliability of certain alerts.

**15. DeviceId**
- Explanation: The unique ID of the device involved in the incident.
- Usefulness: Can be useful if certain devices are more prone to being attacked, but generally not critical for classification.

**16. Sha256**
- Explanation: A hash value that uniquely identifies a file, used for verifying file integrity.
- Usefulness: Could be useful for identifying specific malware if a database of known malicious hashes is available.

**17. IpAddress**
- Explanation: The IP address involved in the incident.
- Usefulness: Important for tracking the origin or target of an attack, especially if combined with geolocation or other network data.

**18. Url**
- Explanation: The URL involved in the incident, often used in phishing or command-and-control operations.
- Usefulness: Very useful when dealing with web-based threats.

**19. AccountSid**
- Explanation: The security identifier for the account involved in the incident.
- Usefulness: Useful for tracking specific accounts, especially if some are more likely to be targeted.

**20. AccountUpn**
- Explanation: The User Principal Name (often an email) of the account involved.
- Usefulness: Can be useful when certain users are more prone to being attacked.

**21. AccountObjectId**
- Explanation: The directory object ID for the account, often used in systems like Active Directory.
- Usefulness: May help in identifying specific users or accounts involved in incidents.

**22. AccountName**

- Explanation: The username associated with the account involved in the incident.
- Usefulness: Similar to AccountUpn, could be used to identify patterns in which users are targeted.

**23. DeviceName**
- Explanation: The name of the device involved in the incident.
- Usefulness: Could be useful if certain devices are more likely to be involved in incidents.

**24. NetworkMessageId**
- Explanation: A unique ID for network messages, likely related to email or other communications.
- Usefulness: Important when tracking communication patterns, especially in phishing attacks.

**25. EmailClusterId**
- Explanation: Groups related emails, possibly indicating a phishing campaign.
- Usefulness: Very useful for identifying and analyzing campaigns rather than individual incidents.

**26. RegistryKey** ,**27. RegistryValueName**, **28. RegistryValueData**

**31. OAuthApplicationId**

- Usefulness: Important when the analysis involves OAuth-related security incidents.

**32. ThreatFamily**
- Explanation: Indicates the family of the threat (e.g., a type of malware).

**33. FileName**

# ***Data Cleaning***

 **Droping the unnessary columns** 

* Columns having huge number of Missing values 

>ActionGrouped,SuspicionLevel,LastVerdict,Roles,ResourceType,ThreatFamily,EmailClusterId,ActionGrouped,ActionGranular

* Columns Understanding is not Clear 

>  ResourceNameId




```python
Train_dataset.drop(['Id','ActionGrouped', 'SuspicionLevel','AntispamDirection', 'LastVerdict', 'Roles', 'ResourceType', 'ThreatFamily', 'EmailClusterId', 'ActionGranular','ResourceIdName'],axis=1,inplace=True)
Train_dataset.head()
```

**Datatype convertion**

Since all the columns are says about the id and unique identifiers of the incidents and There is no need for the mathematical claculation or Mathematical Analysis between the columns I was converting all the columns to the Object datatype Only the Timestampes need to be extracted and in the different datatype

```python
Train_dataset = Train_dataset.astype('object')
Train_dataset.info()
```

**Combining the Columns having similar meaning and contining the grouped information**

***Group 1***

 - AccountSid , AccountUpn , Account Object Id, Account Name ---> Account_Info ====> Used for identifying specific users or accounts involved in incidents.
 
***Group 2***

 - DeviceId,Device Name,Network Message Id ----> Device_Info =====> Used for identifying specific Device involved in incidents.
 
***Group 3***
 
 - RegistoryKey,RegistoryValueData,RegistoryValueName -------> Registory_Info ======> Indicates that the data involves registry keys and values.
 
***Group 4***
 
 - ApplicationId,ApplicationName ------> Application_Info ======> Refers to the applications identified by their unique IDs and names.
 
***Group 5***
 
 - OsFamily,OsVersion -------> OS_Info =======> Refers to the OS configuration of the system.
 
***Group 6***
 
 - State,City,CountryCode -------> Location_info =========> Refers to the Location where the incident happened.
 
***Group 7***
 
 - Ipaddress,Url --------> Device_Ip ========> Refers to the Device Locations with Ip and Url
 
 

**GROUP 1**

```python
print(Train_dataset['AccountName'] .dtype)
```

```python
Train_dataset['Account_Info'] = (Train_dataset['AccountSid'].astype(str) + ' ' +Train_dataset['AccountUpn'].astype(str) + ' ' +Train_dataset['AccountObjectId'].astype(str) + ' ' +Train_dataset['AccountName'].astype(str))
```

```python
Train_dataset.head()
```

**GROUP 2**

```python
Train_dataset['Device_Info'] = (Train_dataset['DeviceId'].astype(str) + ' ' +
                                Train_dataset['DeviceName'].astype(str) + ' ' +
                                Train_dataset['NetworkMessageId'].astype(str))
```

**GROUP 3**

```python
Train_dataset['Registry_Info'] = (Train_dataset['RegistryKey'].astype(str) + ' ' +
                                  Train_dataset['RegistryValueName'].astype(str) + ' ' +
                                  Train_dataset['RegistryValueData'].astype(str))
```

**GROUP 4**

```python
Train_dataset['Application_Info'] = (Train_dataset['ApplicationId'].astype(str) + ' ' +
                                     Train_dataset['ApplicationName'].astype(str))
```

**GROUP 5**

```python
#OsFamily,OsVersion

Train_dataset['OS_Info'] = (Train_dataset['OSFamily'].astype(str) + ' ' +
                            Train_dataset['OSVersion'].astype(str))
```

**GROUP 6**

```python
#State,City,CountryCode
Train_dataset['Location_Info'] = (Train_dataset['State'].astype(str) + ' ' +
                                  Train_dataset['City'].astype(str) + ' ' +
                                  Train_dataset['CountryCode'].astype(str))
```

**GROUP 7**

```python
#Ipaddress,Url
# Concatenate 'Ipaddress' and 'Url' into 'Device_Ip'
Train_dataset['Device_Ip'] = (Train_dataset['IpAddress'].astype(str) + ' ' +
                              Train_dataset['Url'].astype(str))
```

```python
#File Name and FolderPath
Train_dataset['File_Info'] = (Train_dataset['FileName'].astype(str) + ' ' +
                              Train_dataset['FolderPath'].astype(str))
```

**Now droping the Old columns**

```python
Train_dataset.drop(['Url','IpAddress','CountryCode','City','FileName','FolderPath','State','OSVersion','OSFamily','ApplicationName','ApplicationId','RegistryValueData','RegistryValueName','RegistryKey','NetworkMessageId','DeviceName','DeviceId'],axis=1,inplace=True)
```

```python
Train_dataset.head()
```

**Filling the Nullvalues in MitreTechniques since it have a lesser nullvalues**

```python
Train_dataset['MitreTechniques'] = Train_dataset.groupby('Category')['MitreTechniques'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown"))
```

```python
Train_dataset.head()
```

**Checking whether there is any null values is present**

```python
Train_dataset.isnull().sum()/len(Train_dataset)*100
```

Only in the IncidentGrade there is a Null values Since it was the the Target values and the missing value is only 0.5% so I am dropping it.

```python
Train_dataset = Train_dataset.dropna(subset=['IncidentGrade'])
```

```python
Train_dataset.isnull().sum()
```

**Now Extract the Time and Date from the TimeStamp**

```python
# Convert the 'Timestamp' column to datetime
Train_dataset['Timestamp'] = pd.to_datetime(Train_dataset['Timestamp'])

# Extract time, day, month, and year into new columns
Train_dataset['Time'] = Train_dataset['Timestamp'].dt.strftime('%H:%M:%S')  # Extract time in HH:MM:SS format
Train_dataset['Day'] = Train_dataset['Timestamp'].dt.day
Train_dataset['Month'] = Train_dataset['Timestamp'].dt.month
Train_dataset['Year'] = Train_dataset['Timestamp'].dt.year
```

```python
Train_dataset.head()
```

```python
Train_dataset['Time']
```

**Droping the TimeStamp **

```python
Train_dataset.drop(['Timestamp'],axis=1,inplace=True)
```

**Based on time Giving the name** 

```python
Train_dataset['Time'] = pd.to_datetime(Train_dataset['Time'], format='%H:%M:%S').dt.time

# Extract the hour from the time column
Train_dataset['Hour'] = Train_dataset['Time'].apply(lambda x: x.hour)

# Define conditions for categorization
conditions = [
    (Train_dataset['Hour'] < 6),
    (Train_dataset['Hour'] >= 6) & (Train_dataset['Hour'] < 10),
    (Train_dataset['Hour'] >= 10) & (Train_dataset['Hour'] < 16),
    (Train_dataset['Hour'] >= 16) & (Train_dataset['Hour'] < 19),
    (Train_dataset['Hour'] >= 19)
]

# Define corresponding labels
labels = ['Early Morning', 'Morning', 'Afternoon', 'Evening', 'Night']

# Use np.select for vectorized categorization
Train_dataset['PartOfDay'] = pd.Series(pd.cut(Train_dataset['Hour'], 
                                               bins=[-1, 5, 9, 15, 18, 23], 
                                               labels=labels))

# Display the updated DataFrame
print(Train_dataset[['Time', 'PartOfDay']])
```

Droping the Hour and time

```python
Train_dataset.drop(['Hour','Time','AccountSid ','AccountUpn','AccountObjectId','AccountName'],axis=1,inplace=True)
```
